## Matplotlib

Matplot lib is a powerful 2D plotting library for producing publication quality figures. Below we will demonstrate some of it's key features.

First, we will load the decay data to plot using pandas:

```python
import pandas as pd

decaydata = pd.read_csv('data/decays.csv', sep=",")
decaydata.head(n = 5)
```

The spaces and special characters in the column namea may cause us trouble, so let's change these. Let's rename these:

```python
decaydata.columns = [['time', 'decays']]
decaydata.head(n = 5)
```

Nexy we will start to plot using the Pyplot module of the Matplotlib. To do this we will import the module, and if we are working in the notebook we and we want to have the plots show up in the notebook then we need to make this explcit:

```python
# load library
import matplotlib.pyplot as plt 

# make plots show up in ipython notebook
%matplotlib inline
```

To make a simple 2D line plot, we use the `plot` function and pass the x and y values:

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays)

# show plot
plt.show()
```

The data look great, but we have no axis labels... Let's fix this!

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays)

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# show plot
plt.show()
```

Another thing we might want to change are the axis ranges. For example, if we want to display the x-axis only up until 5s (nothing seems to happen after this...) then we can do the following:

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays)

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(0, 5)

# show plot
plt.show()
```

To change the y-axis, you would use `ylim`

Perhaps now, we were interested in making a scatter plot instead of a line plot. We can choose a plotting character, color and linestyle and add that as an argument to the `plot` command. If you want to know all the possible plotting characters, visit this website: http://matplotlib.org/api/markers_api.html

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays,  marker='o', color='r', linestyle='') # specifies marker and linestyle

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(-0.25, 5.5) # also modified xlim to make sure we can see all of the points
plt.ylim(-0.25, 10.5) # also added ylim to make sure we can see all the points

# show plot
plt.show()
```

We can also add text to our plot, for example we can add a title:

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays,  marker='o', color='r', linestyle='') # specifies marker and linestyle

# add plot title
plt.title('Radioactive Decay')

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(-0.25, 5.5) # also modified xlim to make sure we can see all of the points
plt.ylim(-0.25, 10.5) # also added ylim to make sure we can see all the points

# show plot
plt.show()
```

If we want to know a value on the plot, we might want to add gridlines to make it easier to read. 

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays,  marker='o', color='r', linestyle='') # specifies marker and linestyle

# add gridlines to plot
plt.grid(True)

# add plot title
plt.title('Radioactive Decay')

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(-0.25, 5.5) # also modified xlim to make sure we can see all of the points
plt.ylim(-0.25, 10.5) # also added ylim to make sure we can see all the points

# show plot
plt.show()
```

To make this more readable, I will again modify the color and the linestyle:

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays,  marker='o', color='b', linestyle='--') # specifies marker and linestyle

# add gridlines to plot
plt.grid(True)

# add plot title
plt.title('Radioactive Decay')

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(-0.25, 5.5) # also modified xlim to make sure we can see all of the points
plt.ylim(-0.25, 10.5) # also added ylim to make sure we can see all the points

# show plot
plt.show()
```

Finally, I am happy with the figure and I want to use it for my manuscript. I will now save the figure using the `savefig` command.

```python
# add data to plot
plt.plot(decaydata.time, decaydata.decays,  marker='o', color='b', linestyle='--') # specifies marker and linestyle

# add gridlines to plot
plt.grid(True)

# add plot title
plt.title('Radioactive Decay')

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(-0.25, 5.5) # also modified xlim to make sure we can see all of the points
plt.ylim(-0.25, 10.5) # also added ylim to make sure we can see all the points

# save plot
plt.savefig(filename = 'decay_lineplot.pdf')
```

*Note - it is **VERY** important to remove the `plt.show()` command from your code when you are saving your plot or you will save an empty plot...*

We may or may not like the default plot size that the figures are saved to. To change this, we can use the `figure` function:

```python
# change figure size
plt.figure(figsize=(3,2))

# add data to plot
plt.plot(decaydata.time, decaydata.decays,  marker='o', color='b', linestyle='--') # specifies marker and linestyle

# add gridlines to plot
plt.grid(True)

# add plot title
plt.title('Radioactive Decay')

# add axis labels
plt.ylabel('Decays (#)')
plt.xlabel('Time (s)')

# modify axis value intervals
plt.xlim(-0.25, 5.5) # also modified xlim to make sure we can see all of the points
plt.ylim(-0.25, 10.5) # also added ylim to make sure we can see all the points

# save plot
plt.savefig(filename = 'decay_lineplot.pdf')
```

## Creating mutliple plots

What if we want to create multipanel plots? We can use the `subplot` function. We will load some example data about counts of mosquitos per year.

In this collection of data, each file holds the data for a single location, each row holds the information for a single year at that location, and the columns hold the data on both mosquito numbers and the average temperature and rainfall from the beginning of mosquito breeding season.

```python
site_B2 = pd.read_csv('B2_mosquito_data.csv', sep=',')
site_B2.head(n = 5)
```

```python
site_B2.tail(n = 5)
```

Let's now plot 3 panels in a single plot, with year as the x-axis, and temperature, rainfall and mosquitos as the y-axes, each in a different subplot:

```python
# define subplot 1
plt.subplot(1, 3, 1)

# define subplot 2
plt.subplot(1, 3, 2)

# define subplot 2
plt.subplot(1, 3, 3)
```

They look a little squished... let's add a function to allow the subplots to breath:

```python
# define subplot 1
plt.subplot(1, 3, 1)

# define subplot 2
plt.subplot(1, 3, 2)

# define subplot 2
plt.subplot(1, 3, 3)

# allow plots room to breath
plt.tight_layout()
```

Now let's add the data to each plot

```python
# define subplot 1
plt.subplot(1, 3, 1)
plt.plot(site_B2.year, site_B2.temperature)

# define subplot 2
plt.subplot(1, 3, 2)
plt.plot(site_B2.year, site_B2.rainfall)


# define subplot 2
plt.subplot(1, 3, 3)
plt.plot(site_B2.year, site_B2.mosquitos)


# allow plots room to breath
plt.tight_layout()
```

# Other kinds of plots

We have become pretty comfortable of plotting line and scatter plots, but these are not the only type of plots we might like to make. Box plots, histograms. These and more are all possible with Matplotlib. Documentation on how to create these plots can be found here: http://matplotlib.org/gallery.html

### Histogram

```python
# define subplot 1
plt.subplot(1, 3, 1)
plt.hist(site_B2.mosquitos)

# define subplot 2
plt.subplot(1, 3, 2)
plt.hist(site_B2.temperature)


# define subplot 2
plt.subplot(1, 3, 3)
plt.hist(site_B2.rainfall)


# allow plots room to breath
plt.tight_layout()

plt.show()
```

### Boxplot

```python
# define subplot 1
plt.subplot(1, 3, 1)
plt.boxplot(site_B2.mosquitos)

# define subplot 2
plt.subplot(1, 3, 2)
plt.boxplot(site_B2.temperature)

# define subplot 2
plt.subplot(1, 3, 3)
plt.boxplot(site_B2.rainfall)

# allow plots room to breath
plt.tight_layout()

plt.show()
```

### Violinplot

```python
# define subplot 1
plt.subplot(1, 3, 1)
plt.violinplot(site_B2.mosquitos)

# define subplot 2
plt.subplot(1, 3, 2)
plt.violinplot(site_B2.temperature)

# define subplot 2
plt.subplot(1, 3, 3)
plt.violinplot(site_B2.rainfall)

plt.show()
```

## Comparing multiple groups

These plots tell us something about the distribution of **all** measurements, but what about measurements per decade? To answer that question we are best to lean on pandas. Remember, when we are plotting in pandas, since pandas sits on top of Matplotlib we can use the matplotlib functions as well.

First we make a column which will allow us to group by decade:

```python
site_B2['decade'] = site_B2['year'] // 10 * 10
site_B2.head(n = 5)
```

Now we can plot and tell boxplot what to plot by:

```python
site_B2[['decade', 'mosquitos']].boxplot(by='decade')
plt.show()
```

This also works for violing plots:

```python
site_B2.hist(column='mosquitos', by='decade')

plt.show()
```

We get an error becuase there is only one value for 2010, so let's remove that from what we plot:

```python
site_B2[site_B2.decade < 2010].hist(column='mosquitos', by='decade')

plt.show()
```

Works, but it's kind of squished! Normally `plt.tight_layout()` would work, but that's not the case when doing histogram subplots with pandas. Looking at the [documentation](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.hist.html) we see a `figsize` argument, let's try that:

```python
site_B2[site_B2.decade < 2010].hist(column='mosquitos', by='decade', figsize=(5, 7))

plt.show()
```

So how do we know what works and what doesn't? Here's a few tips:

1. Read the docs
2. Try things out

# Making a fancy poster with Matplotlib!

```python
# Import various necessary Python and matplotlib packages
import numpy as np
import matplotlib.cm as cm
from matplotlib.pyplot import figure, show, rc
from matplotlib.patches import Ellipse

# Create a square figure on which to place the plot
fig = figure(figsize=(8,8))

# Create square axes to hold the circular polar plot
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)

# Generate 20 colored, angular wedges for the polar plot
N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10*np.random.rand(N)
width = np.pi/4*np.random.rand(N)
bars = ax.bar(theta, radii, width=width, bottom=0.0)
for r,bar in zip(radii, bars):
    bar.set_facecolor(cm.jet(r/10.))
    bar.set_alpha(0.5)

# Using dictionaries, create a color scheme for the text boxes
bbox_args = dict(boxstyle="round, pad=0.9", fc="green", alpha=0.5)
bbox_white = dict(boxstyle="round, pad=0.9", fc="1", alpha=0.9)
patch_white = dict(boxstyle="round, pad=1", fc="1", ec="1")

# Create various boxes with text annotations in them at specific
# x and y coordinates
ax.annotate(" ",
    xy=(.5,.93),
    xycoords='figure fraction',
    ha="center", va="center",
    bbox=patch_white)

ax.annotate('Matplotlib and the Python Ecosystem for Scientific Computing',
    xy=(.5,.95),
    xycoords='figure fraction',
    xytext=(0, 0), textcoords='offset points',
    size=15,
    ha="center", va="center",
    bbox=bbox_args)

ax.annotate('Author and Lead Developer \n of Matplotlib ',
    xy=(.5,.82),
    xycoords='figure fraction',
    xytext=(0, 0), textcoords='offset points',
    ha="center", va="center",
    bbox=bbox_args)

ax.annotate('John D. Hunter',
    xy=(.5,.89),
    xycoords='figure fraction',
    xytext=(0, 0), textcoords='offset points',
    size=15,
    ha="center", va="center",
    bbox=bbox_white)

ax.annotate('Friday November 5th  \n 2:00 pm \n1106ME ',
    xy=(.5,.25),
    xycoords='figure fraction',
    xytext=(0, 0), textcoords='offset points',
    size=15,
    ha="center", va="center",
    bbox=bbox_args)

ax.annotate('Sponsored by: \n The Hacker Within, \n'
    'The University Lectures Committee, \n The Department of '
    'Medical Physics\n and \n The American Nuclear Society',
    xy=(.78,.1),
    xycoords='figure fraction',
    xytext=(0, 0), textcoords='offset points',
    size=9,
    ha="center", va="center",
    bbox=bbox_args)

#fig.savefig("plot.pdf")
```

## Bokeh

```python
import numpy as np
# import the Bokeh plotting tools
from bokeh import plotting as bp

# as in the matplotlib example, load decays.csv into a NumPy array
decaydata = np.loadtxt('data/decays.csv',delimiter=",",skiprows=1)

# provide handles for the x and y columns
time = decaydata[:,0]
decays = decaydata[:,1]

# define some output file metadata
bp.output_file("decays.html", title="Experiment 1 Radioactivity")

# create a figure with fun Internet-friendly features (optional)
bp.figure(tools="pan,wheel_zoom,box_zoom,reset,previewsave")

# on that figure, create a line plot
bp.line(time, decays, x_axis_label="Time (s)", y_axis_label="Decays (#)",
     color='#1F78B4', legend='Decays per second')

# additional customization to the figure can be specified separately
bp.curplot().title = "Decays"
bp.grid().grid_line_alpha=0.3

# open a browser
bp.show()
```
